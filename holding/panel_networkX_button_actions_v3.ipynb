```python
%matplotlib inline

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import networkx as nx

import panel as pn
pn.extension()

import hvplot
import hvplot.networkx as hvnx
```

```python
#def complete_graph(pts=15, link=[(0,1)]):
def complete_graph(pts=15):
    
    g = nx.complete_graph(pts)
    pos = nx.circular_layout(g)
    
    #nx.draw_networkx_edges(g, pos, edgelist=link, 
    #                      width=8, edge_color="tab:red")
    
    return hvnx.draw(g, pos, node_size=1000,node_color="skyblue", with_labels=True)
```

```python
%matplotlib inline

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import networkx as nx

import panel as pn
pn.extension()

import hvplot
import hvplot.networkx as hvnx

import random
```

```python
def color_edge(nodes=5):
    
    g = nx.complete_graph(nodes)
    pos = nx.circular_layout(g)
    
    #g.add_edge(1,2, color='r', weight=7)
    
    nx.draw(g, pos, node_size=1000,node_color="skyblue", with_labels=True)
    
    node1 = random.randint(0,(nodes-1))
    node2 = random.randint(0,(nodes-1))
    
    nx.draw_networkx_edges(g, pos, edgelist=[(node1,node2)],
                       width=8, edge_color="tab:red")
    
    return g
```

```python
color_edge(5);
```

```python
button_highlight = pn.widgets.Button(name='Highlight Link')
button_highlight.on_click(color_edge)
```

```python
interactive = pn.bind(color_edge, button_highlight)
```

```python
app1 = pn.Column('## App-1', button_highlight, color_edge, align="center")
app1
```
